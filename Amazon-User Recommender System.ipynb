```python
#Define the top 5 movies with the least audience
l1=Amazon.T
l1.insert(4848,'Total',Amazon.count())
l1['Total']
```

```python

Amazon.set_index(["user_id"], inplace = True, 
                    append = True, drop = True) 
   
# resetting index 
#data.reset_index(inplace = True) 
```

```python
df=Amazon.fillna(0)
df.shape
```

```python
#Divide the data into training and test data
train_data,test_data=train_test_split(df,test_size=0.25)
```

```python
train_data
```

```python
#Build a recommendation model on training data
User_similarity = pairwise_distances(train_data, metric='cosine')
User_similarity.shape
```

```python
User_similarity_T = pairwise_distances(test_data, metric='cosine')
User_similarity_T.shape
```

```python
mean_user_ratings=train_data.mean(axis=1)[:,np.newaxis]
```

```python
rating_diff=(train_data-mean_user_ratings)
```

```python
user_pred=mean_user_ratings+User_similarity.dot(rating_diff)/np.array([np.abs(User_similarity).sum(axis=1)]).T
```

```python
#Make predictions on the test data
mean_user_ratings_T=test_data.mean(axis=1)[:,np.newaxis]
rating_diff_T=(test_data-mean_user_ratings_T)
user_pred_T=mean_user_ratings_T+User_similarity_T.dot(rating_diff_T)/np.array([np.abs(User_similarity_T).sum(axis=1)]).T
user_pred_T.shape
```

```python
mse_T=mean_squared_error(user_pred_T,test_data!=0)
```

```python
mse=mean_squared_error(user_pred,train_data!=0)
```
